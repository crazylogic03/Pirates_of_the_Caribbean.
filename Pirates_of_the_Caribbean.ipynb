```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
from wordcloud import WordCloud
from collections import Counter
```

```python
# Load the Lady Gaga dataset
url = "https://drive.google.com/uc?export=download&id=1xI9BsI6PSsqM8jns81cAGurYNziI0vPD"
df = pd.read_csv(url)
print(df.head())
```

## 🧼 Data Cleaning & Preprocessing

```python
# Combine all lyrics into one text
all_lyrics = " ".join(df['Lyric'].dropna()).lower()

# Remove punctuation and non-alphabetic characters
all_lyrics = re.sub(r'[^a-z\s]', '', all_lyrics)

# Tokenize
words = all_lyrics.split()
print(f"Total words: {len(words)}")
```

## 📊 Word Frequency Analysis

```python
# Count word frequencies
word_freq = Counter(words)
sorted_words = word_freq.most_common()

# Display top 10 words
sorted_words[:10]
```

## 📉 Zipf’s Law Visualization (Log-Log Plot & WordCloud)

```python
# Prepare rank and frequency data
ranks = np.arange(1, len(sorted_words) + 1)
frequencies = np.array([freq for word, freq in sorted_words])

# Plot
plt.figure(figsize=(10,6))
plt.loglog(ranks, frequencies)
plt.title("Zipf's Law in Lady Gaga Lyrics")
plt.xlabel("Rank (log scale)")
plt.ylabel("Frequency (log scale)")
plt.grid(True)
plt.show()
```

```python
text = " ".join(df["Lyric"].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# **🔍 Interpretation**  
We analyzed Lady Gaga's lyrics to test **Zipf's Law**—the rule that a word's frequency is inversely proportional to its rank (e.g., the #1 word appears twice as often as #2). Here's the tea:  

---

### **1. 🎨 Word Cloud: Gaga's Linguistic Signature**  
If her lyrics were a word cloud, it'd scream:  
- **"love," "baby," "dance," "time," "oh"** (so many *oh-oh-ohs!*)  
- **Thematic gems**: "fame," "illusion," "monster" (very *ARTPOP* of her).  

**What this means**:  
✔️ Pop lyrics thrive on repetition and emotional hooks.  
✔️ High-frequency words align with her themes (identity, glamour, heartbreak).  

---

### **2. 📉 The Log-Log Plot: Math Meets Music**  
We plotted word ranks vs. frequencies (log scale) and asked: *Does this look like Zipf's Law?*  

**Expected**: A straight line with slope ≈ **-1**.  
**Found**:  
✅ **Mostly linear**—common words "love" dominate as predicted.  
⚠️ **Bumps in the curve** because:  
   - **Hooks**: "P-p-p-poker face" skews counts.  
   - **Ad-libs**: "Rah-rah-ah-ah-ah" isn't your average vocabulary.  

*(Translation: Gaga follows the rule… but bends it with artistic flair.)*  

---

### **3. 🎤 Conclusion: Zipf's Law, But Make It Fashion**  
✔️ **Yes**, her lyrics obey Zipf's Law - like all natural language.  
🎭 **But** with *pop music exceptions*:  
   - Repetition = inflated word counts.  
   - Sound effects ("oh," "yeah") act like linguistic glitter.  

**Bigger Picture**:  
- Even in creative domains, math sneaks in.  
- Gaga's quirks remind us: **artists play by their own rules** (while still nodding to science).
